```python
# Reading dataset
housing = pd.read_csv("data/housing.csv")
housing_bkp = housing.copy()
housing.head()
```

# Understanding about data 

```python
# Numerical variables
num_vars = ["area","bedrooms","bathrooms","stories","parking","price"]
```

```python
df_num = housing[num_vars]
```

```python
# Check the numerical variables in scatter plots
plt.figure(figsize=(10,10))
sns.pairplot(df_num)
plt.show()
```

- area is the best numerical variable - But it is having a lot widespread
- It is useless to compare it with other variables  using a scatter plot

```python
df_num.describe()
```

- Except area and price, other variables are almost categorical 

```python
# To plot clear view of area vs price
plt.figure(figsize=(8,8))
sns.scatterplot(x="area",y="price",data=df_num)
plt.show()
```

- In general the price increases with area, but after certain point, it do not increase
- It may be certainly because of some other dominant variable

```python
# Let us study the effects of each variable except area
num_cats = ["bedrooms","bathrooms","stories","parking"]
# what are distributions of each variables?
fig, axes = plt.subplots(4,figsize=(10,15))

for i,j in enumerate(num_cats):
    sns.distplot(df_num[j],ax=axes[i])

plt.show()
```

```python
# How are the related to price
fig,axis = plt.subplots(4,figsize=(10,15))
for i, j in enumerate(num_cats):
    sns.boxplot(x=j,y="price",data=df_num, ax=axis[i])
plt.show()
```

- For Bedrooms, prices increases gradually as we increase the value. But 6 bedroom flats are not that much priced
- Increase in number of Bathrooms clearly increasing the price
- similar for stories - 3 stories and 4 stories cost significantly higher than 1st and 2nd story buildings
- Buildings having more parking spaces definitely priced higher for 1 or 2. But not for 3 spaces - none cares

```python
# Let us analyse the Categorical ( Characters)  variables
cat_vars = ["mainroad","guestroom","airconditioning","basement","hotwaterheating","airconditioning","prefarea","furnishingstatus"]
len(cat_vars)
```

```python
# Understanding the effects of each variables with price
fig,axes = plt.subplots(8,figsize=(10,30))

for i,j in enumerate(cat_vars):
    sns.boxplot(x=j,y="price",data=housing,ax = axes[i])

plt.show()
```

##### Let us list down our findings 
- first of all, Mainroad - housing near to main road is significantly higher 
- Having a guestroom matters 
- If you need airconditioned, pay more price
- If house within prefarea, the price increases
- At last, Unfurnished ones cost less than semi-furnished and furnished ones.

```python
for i in cat_vars:
    print("The number of items in each Category for : ",i)
    print(housing[i].value_counts())
```

```python
varlist=["mainroad","guestroom","airconditioning","basement","hotwaterheating","prefarea"]
housing[varlist] = housing[varlist].apply(lambda x: x.map({"yes":1, "no":0}))
housing.head()
```

- Mark my words, Mulitple Linear Regression wont give r2 more than 70%

```python
# Lets create dummy variables for furnishing status
stat = pd.get_dummies(housing["furnishingstatus"])
stat.head()
```

```python
# Let us attach this to original dataframe - keep furnishing status column aside for the moment.
housing = pd.concat([housing,stat],axis=1)
housing_lm = housing.drop("furnishingstatus",axis=1)
housing_lm.head()
```

```python
housing2 = housing_lm.copy()
```

```python
# Performing normalization over dataset before modelling
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
housing_lm[num_vars] = scaler.fit_transform(housing_lm[num_vars])
housing_lm.head()
```

```python
# Fine Now split training and test set
from sklearn.model_selection import train_test_split
```

```python
df_train_lm, df_test_lm = train_test_split(housing_lm,train_size=0.7,random_state=100)
```

```python
y_train_lm = df_train_lm["price"]
X_train_lm = df_train_lm.drop("price",axis=1)
```

```python
# Building first model with all the variables 
import statsmodels.api as sm
```

```python
X_train_lm_sm = sm.add_constant(X_train_lm)
mod = sm.OLS(y_train_lm,X_train_lm_sm)
linearmodel1 = mod.fit()
```

```python
linearmodel1.summary()
```

- Pathetic little linear model

- Calculating VIF  for multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
X_train_lm_sm.shape[1]
```

```python
vif = [variance_inflation_factor(X_train_lm_sm.values,i) for i in range(X_train_lm_sm.shape[1])]
```

```python
df_vif = pd.DataFrame(vif,index=X_train_lm_sm.columns)
df_vif
```

- Oops, I must had removed one dummy variable under furnishing status

```python
X_train_lm_sm = X_train_lm_sm.drop("furnished",axis=1)
vif = [variance_inflation_factor(X_train_lm_sm.values,i) for i in range(X_train_lm_sm.shape[1])]
df_vif = pd.DataFrame(vif,index=X_train_lm_sm.columns)
df_vif
```

```python
# Lets make new model 
mod = sm.OLS(y_train_lm,X_train_lm_sm)
linearmodel2 = mod.fit()
linearmodel2.summary()
```

- Removing high p -values 

```python
X_train_lm_sm = X_train_lm_sm.drop(["bedrooms","semi-furnished"],axis=1)
mod = sm.OLS(y_train_lm,X_train_lm_sm)
linearmodel2 = mod.fit()
linearmodel2.summary()
```

- This is ok but r2 is only 68% 

```python
# Let us predict against the training dataset
y_train_lm_pred = linearmodel2.predict(X_train_lm_sm)
```

# Residual analysis

```python
plt.figure(figsize=(10,8))
sns.distplot((y_train_lm-y_train_lm_pred))
plt.show()
```

- Distribution is almost normal wrt 0

```python
# Evaluation against Test Set
y_test_lm = df_test_lm["price"]
X_test_lm = df_test_lm[X_train_lm_sm.drop("const",axis=1).columns]
```

```python
X_test_lm_sm = sm.add_constant(X_test_lm)
y_test_lm_pred = linearmodel2.predict(X_test_lm_sm)
```

```python
plt.figure(figsize=(10,8))
sns.distplot((y_test_lm-y_test_lm_pred))
plt.show()
```

```python
# Calculate r2 for test and train
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
r2_train = r2_score(y_true=y_train_lm,y_pred=y_train_lm_pred)
r2_test = r2_score(y_true=y_test_lm,y_pred=y_test_lm_pred)
print("Training score: ",r2_train)
print("Test Score :",r2_test)
```

```python
mae_train = mean_absolute_error(y_true=y_train_lm,y_pred=y_train_lm_pred)
mae_test = mean_absolute_error(y_true=y_test_lm,y_pred=y_test_lm_pred)
print("Training score: ",mae_train)
print("Test Score :",mae_test)
```

- it is ok - We can try ensemble based algorithms to work on this in future
